```python
import numpy as np
import matplotlib.pyplot as plt
import random
```

```python
def tracking(k):
    x_0 = random.uniform(0, 2*3.14)
    p_0 = random.uniform(0, 2*3.14)
    n=1000
    a=[[x_0,p_0]]

    for time in range (1,n):
        p=p_0+k*np.sin(x_0)
        p= p % (2*3.14)
        x=p+x_0
        x= x % (2*3.14)
        p= p % (2*3.14)
        a.append([x,p])
        x_0=x
        p_0=p
    a = np.array(a)
    return a
```

```python
plt.scatter(a[:,0],a[:,1])
plt.show()
```

```python
k=random.uniform(0, 0.1)
h= np.array([tracking(3) for i in range(1000) ])
```

```python
for i in range(1000):
    plt.scatter(h[i,:,0],h[i,:,1], s=1)
```
