```python
def Kitaev(t1,delta,T,Max):
    print(f'T1={t1},delta={delta},N={T}, Max={Max}')
    E=[]
    K=[]
    N=4*4*T
    mu=0
    t1=t1/2
    delta=delta/2
    Numax=500
    E=[]
    K=[]
    for x in range(Numax):
        mu=x*2*Max/Numax-Max  
        M= np.zeros( (N,N) )
        for i in range(0,N):
            for j in range(0,N,2):
                if i+2==j:
                    M[i][j]=-t1
                    M[j][i]=-t1
            for j in range(1,N,2):
                if i+2==j:
                    M[i][j]=t1
                    M[j][i]=t1
            for j in range(0,N,2):
                if i+1==j:
                    M[i][j]=-delta
                    M[j][i]=-delta
        for i in range(0,N,2):            
            for j in range(0,N,1):
                if i+1==j-2:
                    M[i][j]=delta
                    M[j][i]=delta
        for i in range(0,N,2):
            M[i][i]=-mu
            M[i+1][i+1]=mu
        eigenvalues, eigenvectors = np.linalg.eig(M)
        for l in range(N):
            K.append(mu)
            E.append(eigenvalues[l].real)
    return K, E
```

```python
def SSH(t1,t2,omega,mu,T,Max):
    print(f'T1={t1},t2={t2},omega={omega},mu={mu},N={T}, Max={Max}')
    E=[]
    K=[]
    N=4*4*T
    #omega=0.0
    #mu=0
    Numax=500
    E=[]
    K=[]
    for x in range(Numax):
        t1=x*2*Max/Numax-Max  
        M= np.zeros( (N,N) )
        for i in range(0,N):
            for j in range(0,N,2):       
                if i+1==j:
                    M[i][j]=-t2
                    M[j][i]=-t2
            for j in range(1,N,2): 
                if i+1==j:
                    M[i][j]=-t1
                    M[j][i]=-t1
            for i in range(0,N,2):
                    M[i][i]=omega-mu
                    M[i+1][i+1]=-omega-mu
        eigenvalues, eigenvectors = np.linalg.eig(M)
        for l in range(N):
            K.append(t1)
            E.append(eigenvalues[l].real)
    return K, E
```
