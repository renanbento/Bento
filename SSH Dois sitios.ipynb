```python
def SSH_FINITOep(t1,t2,omega,mu,T,Max):
    print(f'V1={t1},V2={t2},\u03B5={omega},\u03BC={mu},N={T}, Max={Max}')
    E=[]
    K=[]
    N=2*T
    #omega=0.0
    #mu=0
    Numax=1000
    E=[]
    K=[]
    n=[]
    for x in range(Numax):
        omega=x*2*Max/Numax-Max  
        M= np.zeros( (N,N) )
        for i in range(0,N):
            for j in range(0,N,2):       
                if i+1==j:
                    M[i][j]=-t2
                    M[j][i]=-t2
            for j in range(1,N,2): 
                if i+1==j:
                    M[i][j]=-t1
                    M[j][i]=-t1
            for i in range(0,N,2):
                    M[i][i]=omega- mu #np.arctan(omega)-np.arctan(mu)
                    M[i+1][i+1]=-omega -mu#-np.arctan(omega)-np.arctan(mu)
        eigenvalues, eigenvectors = np.linalg.eig(M)
        for l in range(N):
            K.append(omega)
            E.append(eigenvalues[l].real)
            n.append(l)
    return K, E
```

```python
def SSH_INFINITO(t1,t2,A,mu,Max):
    print(f'T1={t1},t2={t2},omega={A},mu={mu}, Max={Max}')
    #t1=1
    #t2=1
    #mu=-0.0
    #A=0.0
    Numax=1000
    E=[]
    K=[]
    for x in range(Numax):
        kx=x*np.pi*2*Max/Numax - Max*np.pi
        C=(t1)+(t2)*np.exp(1j*kx/2)
        C1=(t1)+(t2)*np.exp(-1j*kx/2)
        M= np.zeros( (2,2) )
        MU= np.diag([A-mu,-A-mu])
        H= np.diag([C],1)
        HT= np.diag([C1],-1)
        M=(MU+H+HT)
        eigenvalues, eigenvectors = np.linalg.eig(M)
        for l in range(2):
            K.append(kx)
            E.append(eigenvalues[l].real)
    return K, E
```

```python
def SSH_FINITOep(t1,t2,omega,mu,T,Max):
    print(f'V1={t1},V2={t2},\u03B5={omega},\u03BC={mu},N={T}, Max={Max}')
    E=[]
    K=[]
    N=(T)
    #omega=0.0
    #mu=0
    Numax=1000
    E=[]
    K=[]
    n=[]
    A=[]
    B=[]
    ZE=[]# zero state
    Ecut=0.1
    #for x in range(Numax):
        #omega= 0.1 #x*2*Max/Numax-Max  
    M= np.zeros( (N,N) )
    for i in range(0,N):
        for j in range(0,N,2):       
            if i+1==j:
                M[i][j]=-t2
                M[j][i]=-t2
        for j in range(1,N,2): 
            if i+1==j:
                M[i][j]=-t1
                M[j][i]=-t1
        for i in range(0,N,2):
                M[i][i]=omega- mu #np.arctan(omega)-np.arctan(mu)
                M[i+1][i+1]=-omega -mu#-np.arctan(omega)-np.arctan(mu)
    eigenvalues, eigenvectors = np.linalg.eig(M)
    se=sorted(eigenvalues)
    for i in range(len(eigenvalues)):
        #print(eigenvalues[i])
        if abs(eigenvalues[i]) < 0.01:
            estado=eigenvalues[i]
            ZE.append(i)
            print(estado, i, ZE)
    #for l in range(0,N):
    #    print(eigenvectors[l][5])
    #for l in range(0,N):
    #    for c in range(0,N):
    #        print(f'{M[l][c]:^5}', end=" ")
    #    print()
    for l in range(0,N):
        K.append(omega)
        E.append((se[l]))
        n.append(l)
        if len(ZE)==2:
            A.append((eigenvectors[l][ZE[0]])**2)
            B.append((eigenvectors[l][ZE[1]])**2)
        if len(ZE)==1:
            A.append((eigenvectors[l][ZE[0]])**2)
        if len(ZE)==0:
            B.append((eigenvectors[l][2])**2)
            A=np.zeros(N)
    return K, E , n , A, B
```

```python
def spin(N,vec):
    j=int(N/2)
    spmz= np.zeros( (N,N) )
    vecout= np.zeros( (N) )
    vetor=vec
    veout=np.zeros(N)
    result=0
    for i in range(int(N/2)):
        spmz[i][i]=1
        spmz[i+j][i+j]=-1
        #spmz[i+2*j][i+2*j]=-1
        #spmz[i+3*j][i+3*j]=-1
    for i in range(N):
        vecout[i]=vetor[i].conjugate()
    for i in range(N):
        for j in range(N):
            flag=spmz[i][j]*vetor[j]
            veout[i]=flag+veout[i]
    for i in range(N):
        auxi=vecout[i]*veout[i]
        result=result+auxi
    #print(result)
    return result
```

```python
def teste1(t1,t2,v,A,mu,Max):
    print(f'T1={t1},t2={t2},v={v},epsilon={A},mu={mu}, Max={Max}')
    #t1=2.0
    #t2=1.0
    #mu=-0.0
    #A=0.0
    N=4
    Numax=1000
    E=[]
    K=[]
    sz=[]
    for x in range(Numax):
        kx=x*np.pi*2*Max/Numax - Max*np.pi
        C=(t2*np.exp(1j*kx/2) + t1)
        C1=(t1)+(t2)*np.exp(-1j*kx/2)
        M= np.zeros( (N,N) )
        MU= np.diag([A-mu+v,-A-mu+v,A-mu-v,-A-mu-v])
        H= np.diag([C,0,C],1)
        HT= np.diag([C1,0,C1],-1)
        M=(MU+H+HT)
        np.set_printoptions(threshold=np.inf)
        #print(C)
        #for l in range(0,4):
        #    for c in range(0,4):
        #        print(f'{(M[l][c]).real:^5}', end=" ")
        #    print()
        #print(np.set_printoptions(M) )    
        #break
        eigenvalues, eigenvectors = np.linalg.eig(M)
        for l in range(N):
            K.append(kx)
            E.append(eigenvalues[l].real)
            if v != 0:
                sz.append(spin(N,eigenvectors[l].real))
            if v == 0:
                for i in eigenvectors[l]:
                    sz.append(0)
                    #print(sz)
                    break
    return K, E, sz
```
