```python
import numpy as np
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D
import random
```

```python
def tracking(k):
    alpha=1
    x_0 =np.random.random()
    y_0 =np.random.random()
    z_0 =np.random.random()
    n=100
    a=[[x_0,y_0,z_0]]
    for time in range (1,n):
        x=x_0
        y=y_0*np.cos(alpha)-z_0*np.sin(alpha)
        z=y_0*np.sin(alpha)+z_0*np.cos(alpha)
        x1=x_0*np.cos(k*z)- y*np.sin(k*z) 
        y1=x_0*np.sin(k*z)+ y*np.cos(k*z)
        z1=z
        a.append([x1,y1,z1])
        x_0=x1
        y_0=y1
        z_0=z1
    a = np.array(a)
    return a
```

```python
p=tracking(10)
```

```python
plt.scatter(p[:,0],p[:,1])
plt.show()
```

```python
fig = plt.figure()
ax = Axes3D(fig)
ax.scatter(p[:,0],p[:,1],p[:,2])
plt.show()
```

```python
k=random.uniform(0, 0.1)
pn=10
h= np.array([tracking(5.0) for i in range(pn) ])
```
